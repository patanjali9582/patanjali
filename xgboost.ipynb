```python
import numpy as np
import pandas as pd
import xgboost as xgb
from xgboost import XGBClassifier,DMatrix
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
```

```python
train_set = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data', header = None)
```

```python
train_set
```

```python
test_set = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.test' ,skiprows = 1, header = None)
```

```python
test_set
```

```python
col_labels = ['age', 'workclass', 'fnlwgt', 'education', 'education_num',
'marital_status', 'occupation','relationship', 'race', 'sex', 'capital_gain',
'capital_loss', 'hours_per_week', 'native_country', 'wage_class']
```

```python
train_set.columns = col_labels
test_set.columns = col_labels
```

```python
test_set
```

```python
train_set.workclass = train_set.workclass.str.replace(' ', '') 
train_set.education = train_set.education.str.replace(' ', '')
train_set.marital_status = train_set.marital_status.str.replace(' ', '') 
train_set.occupation = train_set.occupation.str.replace(' ', '')
train_set.relationship = train_set.relationship.str.replace(' ', '') 
train_set.native_country = train_set.native_country.str.replace(' ', '')
train_set.race = train_set.race.str.replace(' ', '') 
train_set.sex = train_set.sex.str.replace(' ', '')
```

```python
test_set.workclass = test_set.workclass.str.replace(' ', '') 
test_set.education = test_set.education.str.replace(' ', '')
test_set.marital_status = test_set.marital_status.str.replace(' ', '') 
test_set.occupation = test_set.occupation.str.replace(' ', '')
test_set.relationship = test_set.relationship.str.replace(' ', '') 
test_set.native_country = test_set.native_country.str.replace(' ', '')
test_set.race = test_set.race.str.replace(' ', '') 
test_set.sex = test_set.sex.str.replace(' ', '')
```

```python
train_set = train_set[train_set['workclass']!='?']
train_set = train_set[train_set['occupation']!='?']
train_set = train_set[train_set['native_country']!='?']
```

```python
test_set = test_set[test_set['workclass']!='?']
test_set = test_set[test_set['occupation']!='?']
test_set = test_set[test_set['native_country']!='?']
```

```python
labelencoder = LabelEncoder()
```

```python
train_set['workclass']= labelencoder.fit_transform(train_set['workclass'])

train_set['education']= labelencoder.fit_transform(train_set['education'])

train_set['marital_status']= labelencoder.fit_transform(train_set['marital_status'])

train_set['occupation']= labelencoder.fit_transform(train_set['occupation'])

train_set['relationship']= labelencoder.fit_transform(train_set['relationship'])
train_set['race']= labelencoder.fit_transform(train_set['race'])
train_set['sex']= labelencoder.fit_transform(train_set['sex'])
train_set['native_country']= labelencoder.fit_transform(train_set['native_country'])
train_set['wage_class']= labelencoder.fit_transform(train_set['wage_class'])
```

```python
test_set['workclass']= labelencoder.fit_transform(test_set['workclass'])

test_set['education']= labelencoder.fit_transform(test_set['education'])

test_set['marital_status']= labelencoder.fit_transform(test_set['marital_status'])

test_set['occupation']= labelencoder.fit_transform(test_set['occupation'])

test_set['relationship']= labelencoder.fit_transform(test_set['relationship'])
test_set['race']= labelencoder.fit_transform(test_set['race'])
test_set['sex']= labelencoder.fit_transform(test_set['sex'])
test_set['native_country']= labelencoder.fit_transform(test_set['native_country'])
test_set['wage_class']= labelencoder.fit_transform(test_set['wage_class'])
```

```python
x_train = train_set.drop(columns = ['wage_class'])
```

```python
y_train = train_set['wage_class']
```

```python
x_test = test_set.drop(columns = ['wage_class'])
y_test = test_set['wage_class']
```

```python
model = XGBClassifier()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_train)
predictions = [round(value) for value in y_pred]
accuracy = accuracy_score(y_train,predictions)
accuracy
```

```python
y_pred = model.predict(x_test)
predictions = [round(value) for value in y_pred]
accuracy = accuracy_score(y_test,predictions)
accuracy
```

```python
grid_param={'learning_rate':[1,0.5,0.3,0.1,0.05,0.01],
           'max_depth':[2,3,6,10,15,100,150],
           'n_estimators':[10,20,30,100]}
```

```python
grid= GridSearchCV(XGBClassifier(objective='binary:logistic'),grid_param, verbose=3)
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```
